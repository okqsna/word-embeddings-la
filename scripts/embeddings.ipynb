Embeddings matrices

```python
train_dict = {}

with open("../data/usage/uk-en-train.csv", "r", encoding="utf-8") as f_train:
    next(f_train)
    for line in f_train:
        line = line.strip("\n")
        ua_word, eng_word = line.split(" ")

        if ua_word not in train_dict:
            train_dict.setdefault(ua_word, [])
        train_dict[ua_word].append(eng_word)
```
